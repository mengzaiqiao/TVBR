## Loading dataset

```python
from beta_rec.data.grocery_data import GroceryData
from beta_rec.datasets.dunnhumby import Dunnhumby

dataset = Dunnhumby(min_u_c=30, min_i_c=20, min_o_c=10)
split_dataset = dataset.load_temporal_basket_split(test_rate=0.2)
data = GroceryData(split_dataset)
```

### Model config

```python
config = {
    "config_file": "../configs/tvbr_default.json",
    "device": "gpu",
    "n_sample": 1000000,
}
# the 'config_file' key is required, that is used load a default config.
# Other keys can be specified to r|eplace the default settings.
```

### Model intialization and training

```python
from beta_rec.recommenders import TVBR

model = TVBR(config)
model.train(data)
```

```python
model.test(data.test[0])
```
